### 計算每個多音字中每個拼音在該多音字樣本中的佔筆

```python

import json
import csv
from collections import defaultdict

# 取出所有多音字字元，共623個
unique_chars = set()

with open("./concate_cpp/total.json","r", encoding='utf-8') as f:
    total_data = json.load(f)
    
for entry in total_data:
    polyphone = entry["content"].split("**")[1]
    for match in polyphone:
        match = ''.join([char for char in match if char.isalnum()])
    unique_chars.update(match)
        
unique_char_list = list(unique_chars)
# print("字元數量:", len(unique_char_list))


#計算每個字的每個拼音數量
result = defaultdict(lambda: defaultdict(int))

for entry in total_data:
    if not entry['content'].split("**")[1].isalnum():  # 處理特例
        content = entry['content'].split("**")[2]
    else:
        content = entry['content'].split("**")[1]
    groundtruth = entry['groundtruth']
    
    for char in unique_char_list:
        if char in content:
            result[char][groundtruth] += 1

# 用來計算每個字的拼音比例
final_result = {}

for char, pinyin_count in result.items():
    total_count = sum(pinyin_count.values())
    pinyin_ratio = {pinyin: round(count / total_count, 4) for pinyin, count in pinyin_count.items()}
    final_result[char] = pinyin_ratio

with open("pinyin_ratio.csv", "w", newline='', encoding="utf-8") as file:
    fieldnames = ['char', 'pinyin', 'ratio']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    for char, pinyin_ratio in final_result.items():
        for pinyin, ratio in pinyin_ratio.items():
            writer.writerow({'char': char, 'pinyin': pinyin, 'ratio': ratio})

print("CSV 檔案已經生成。")
```

### 統計cpp資料集中的(字,拼音)不在字典中的組數

```python
import pandas as pd

def check_missing_pinyin():
    dictionary_df = pd.read_csv("dictionary_zdic.csv")
    pinyin_ratio_df = pd.read_csv("pinyin_ratio.csv")
    dictionary_set = set(zip(dictionary_df['character'], dictionary_df['pinyin']))
    
    missing_pairs = []
    count = 0
    for _, row in pinyin_ratio_df.iterrows():
        pair = (row['char'], row['pinyin'])
        if pair not in dictionary_set:
            missing_pairs.append(pair)
            count+=1
            print(f"{count},{pair}")

check_missing_pinyin()
```

### 統計字典中的(字,拼音)不在cpp資料集中的組數

```python
import pandas as pd

def check_pinyin_coverage():
    dictionary_df = pd.read_csv("dictionary_zdic.csv")
    pinyin_ratio_df = pd.read_csv("pinyin_ratio.csv")
    pinyin_ratio_set = set(zip(pinyin_ratio_df['char'], pinyin_ratio_df['pinyin']))

    missing_pairs = []
    count = 0
    # 檢查 dictionary_zdic.csv 中的 (character, pinyin) 是否存在於 pinyin_ratio_set
    for _, row in dictionary_df.iterrows():
        char_pinyin_pair = (row['character'], row['pinyin'])
        if char_pinyin_pair not in pinyin_ratio_set:
            missing_pairs.append(char_pinyin_pair)
            count += 1
            print(f"{count},{char_pinyin_pair}")
        
    return missing_pairs

missing_pairs = check_pinyin_coverage()

```

### 對字典進行清理，處理不在cpp中的(字,拼音)

```python
def remove_missing_pairs(missing_pairs):
    dictionary_df = pd.read_csv("dictionary_zdic.csv")
    missing_pairs_set = set(missing_pairs)
    filtered_df = dictionary_df[~dictionary_df[['character', 'pinyin']].apply(tuple, axis=1).isin(missing_pairs_set)]
    filtered_df.to_csv("cleaned_dictionary.csv", index=False, encoding="utf-8-sig")
    print("清理完成，結果已儲存至 'cleaned_dictionary.csv'")

remove_missing_pairs(missing_pairs)
```
